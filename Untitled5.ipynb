```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import os
import glob
from statistics import mean
```

```python
rubles = {'1': [], '2': [], '5': [], '10': []}

for value in ['1', '2', '5', '10']:
  image_paths = glob.glob(f'/content/drive/My Drive/Colab Notebooks/rubles/{value}_rubles/*.png', recursive=True)

  for img_path in image_paths:
      img = cv2.imread(img_path)
      img = cv2.resize(img, (100, 100))
      cv2_imshow(img)
      rubles[value].append(img)
```

```python
img = cv2.imread('/content/drive/My Drive/Colab Notebooks/rubles/img.png')

img = cv2.resize(img, (600, 400))

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

blurred = cv2.GaussianBlur(gray, (15, 15), 0)
```

```python
circles = cv2.HoughCircles(
        blurred,
        cv2.HOUGH_GRADIENT,
        dp=1.0,
        minDist=50,
        param1=100,
        param2=30,
        minRadius=20,
        maxRadius=150
    )

output = img.copy()
```

```python
masks = {}

for i, (x, y, r) in enumerate(circles[0]):
    x, y, r = int(x), int(y), int(r)
    print(x, y, r)

    cv2.circle(output, (x, y), r, (68, 255, 22), 4)

    img_coins = img.copy()

    mask = np.zeros(img_coins.shape[:2], dtype="uint8")
    cv2.circle(mask, (x, y), r, (68, 255, 22), -1)
    masked = cv2.bitwise_and(img_coins, img_coins, mask=mask)

    # save image
    cv2.imwrite(f'/content/drive/My Drive/Colab Notebooks/rubles/coins/coins{i + 1}.png', masked)

    masks[i + 1] = mask
```

```python
recognized = []
for i in range(len(masks.keys())):
  img = cv2.imread(f'/content/drive/My Drive/Colab Notebooks/rubles/coins/coins{i+1}.png')
  # img = cv2.resize(img, (300, 200))
  cv2_imshow(img)
  recognized.append(img)
```
